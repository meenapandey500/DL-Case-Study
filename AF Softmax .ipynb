```python
#Example for softmax activation function
def softmax(X):
    a = np.exp(X)
    total = np.sum(a)
    return a/total
```

```python
import numpy as np
```

```python
#main program
#We have input X
X= np.array([20,15,5])
#call function
result=softmax(X)
print(result)
for i in result:
  print(round(i,2))

print("sum : ",np.sum(result))
```
